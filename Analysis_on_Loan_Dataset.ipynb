```python
df["Gender"]=df["Gender"].fillna(df["Gender"].mode()[0])
```

```python
df["LoanAmount"].median()
```

```python
df["LoanAmount"]=df["LoanAmount"].fillna(df["LoanAmount"].median())
```

```python
df["Total_Income"]=df["ApplicantIncome"]+df["CoapplicantIncome"]
```

```python
df["Ratio"]=df["LoanAmount"]/df["Total_Income"]*100
```

```python
df["Ratio"].describe()
```

```python
def income_grp(pi):
    if pi<5000:
        group="Low Income"
    elif pi>5000 and pi<10000:
        group="Mid Income"
    else :
        group="High Income"
    return group
```

```python
df["Binning"]=df["Total_Income"].apply(income_grp)
```

```python
def grp(row):
    if(row["Married"]=="Yes") and (row["Education"]=="Graduate") and (row["Self_Employed"]=="No"):
       group="stable"
    else:
       group="Unstable"
    return group
```

```python
df["group"]=df.apply(grp,axis=1)
```

```python
df.to_csv("Updated_loan.csv")
```
